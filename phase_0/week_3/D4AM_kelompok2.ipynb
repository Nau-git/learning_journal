**H0: μ <= \$5.948**

**H1: μ > \$5.948**

```python
from scipy import stats
t_stat,p_val = stats.ttest_1samp(USA_df, usa_mean)
print('P-value:',p_val/2) #The p-value divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat)
```

```python
from scipy import stats
t_stat2,p_val2 = stats.ttest_1samp(df.Total, 10)
print('P-value:',p_val2/2) #The p-value divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat2)
```

```python
uk_mean = df[(df['Country'] == "USA")]['Total'].mean()
uk_mean
```

```python
daily_uk = df[df['Country']=='United Kingdom'][['InvoiceDate','Total']].groupby('InvoiceDate').sum()
daily_usa = df[df['Country']=='USA'][['InvoiceDate','Total']].groupby('InvoiceDate').sum()

print('Average sales of United Kingdom a day: ${}'.format(np.round(daily_uk['Total'].mean())))
print('Average sales of USA a day: ${}'.format(np.round(daily_usa['Total'].mean())))
```

```python
t_stat, p_val = stats.ttest_ind(daily_uk,daily_usa)
print('P-value:',p_val[0]) 
print('t-statistics:',t_stat[0])
```

```python
uk_pop = np.random.normal(daily_uk['Total'].mean(),daily_uk['Total'].std(),10000)
us_pop = np.random.normal(daily_usa['Total'].mean(),daily_usa['Total'].std(),10000)

ci = stats.norm.interval(0.95, daily_uk['Total'].mean(), daily_uk['Total'].std())
plt.figure(figsize=(16,5))
sns.distplot(uk_pop, label='UK Average Sales a Day *Pop',color='blue')
sns.distplot(us_pop, label='us Average Sales a Day *Pop',color='red')

plt.axvline(daily_uk['Total'].mean(), color='blue', linewidth=2, label='UK mean')
```
