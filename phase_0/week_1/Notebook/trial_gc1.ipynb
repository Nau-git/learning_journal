#### Grouping

```python
poke.tail()
```

```python
poke.groupby('Type_1')['Speed'].mean()
```

```python
poke.groupby('Type_2')['Speed'].mean()
```

```python
poke[
        (poke['Type_1'] == 'Normal') &
        (poke['Speed'] >= 100)
].groupby(['Generation', 'Speed'])['Name'].count()
# Pokemon di tiap Generation dgn Type_1 Normal dan Speed >= 100
```

```python
poke[
        ((poke['Type_1'] == 'Fire') |
        (poke['Type_2'] == 'Fire')) &
        (poke['Speed'] >= 100)
].groupby(['Legendary', 'Speed'])['Name'].count()
# Pokemon Legendary yg punya atribut Fire dengan Speed >= 100
```

```python
poke[
        ((poke['Type_1'] == 'Fire') |
        (poke['Type_2'] == 'Fire')) &
        (poke['Speed'] >= 100)
].groupby(['Generation', 'Speed'])['Name'].count()
# Pokemon yg punya atribut Fire dengan Speed >= 100 di tiap generasi
```

```python
poke[
        (poke['Total'] >= 600)
].groupby(['Generation', 'Name'])['#'].count()
# Pokemon terkuat Total >= 600 di tiap generasi
```
