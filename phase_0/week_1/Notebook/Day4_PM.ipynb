#### **Query**

```python
df.head()
```

```python
df[df['year_id'] > 2010]
```

```python
# year_id > 2010
# fran_id == Warriors
# pts > 100
# game_result == W

df[
    (df['year_id']>2010) &
    (df['fran_id']== "Warriors") &
    (df['pts']>100) &
    (df['game_result']== 'W')

]         # ini 1 query
```

```python
# bukan query?
df[
    (df['year_id']>2010) &
    (df['fran_id']== "Warriors") &
    (df['pts']>100) &
    (df['game_result']== 'W')
][['year_id', 'fran_id', 'pts']] # cuma nampilin beberapa kolom
```

```python
# penulisan dibalik. Specify column dulu, baru query
df[['year_id', 'fran_id', 'pts']] [
    (df['year_id']>2010) &
    (df['fran_id']== "Warriors") &
    (df['pts']>100) &
    (df['game_result']== 'W')
]
```

```python
# alternatif penulisan query

# year_id > 2010
# fran_id == Warriors
# pts > 100
# game_result == W

df.query('year_id > 2010 & fran_id == "Warriors" & pts > 100 & game_result == "W"')  # ini 1 query
```

```python
# bisa juga pilih kolom yg mau ditampilin
df.query('year_id > 2010 & fran_id == "Warriors" & pts > 100 & game_result == "W"')[['year_id', 'fran_id', 'pts']]
```

```python
# bukan query
df[
    (df['year_id']>2010) &
    (df['fran_id']== "Warriors") &
    (df['pts']>100) &
    (df['game_result']== 'W')
][['year_id', 'fran_id', 'pts']]
```

# Grouping n Aggregating

```python
df.head()
```

```python
df.groupby('fran_id')['pts'].mean()
```

```python
df.groupby('game_location')['pts'].mean()  # (grouping)[value yang diingin].Aggregating
```

cara kerjanya:
* 1. grouping -> game_location
* 2. aggregate -> mean pada pts

detailnya:

```python
# 1. grouping
grup = df.groupby('game_location')
grup
```

```python
# 2a. aggregate
grup.get_group('A')['pts'].mean()
```

```python
# 2b. aggregate
grup.get_group('H')['pts'].mean()
```

```python
# 2c. aggregate
grup.get_group('N')['pts'].mean()
```

Bandingkan semua hasilnya. Sama semua

```python
# chain query + groupby

df[
    (df['year_id'] > 2012) &
    (df['fran_id'] == 'Spurs') &
    (df['game_result'] == 'W')
].groupby(['year_id', 'game_location'])['pts'].mean()
```

```python
# fran dgn rata2 pts > 100

df.groupby('fran_id')['pts'].mean()
```

```python
# fran dgn rata2 pts > 100

df.groupby('fran_id')['pts'].mean().sort_values(ascending=False)  # Series
```

```python
df2 = df.groupby('fran_id')[['pts']].mean()
df2[df2['pts']>100]  # DataFrame
```

# Manipulating Columns
* create new column

* rename

* drop

```python
df_new = df.copy()  #bikin backup karena akan merubah data
df_new.head()
```

```python
# Create new column
df_new['selisih_poin'] = df_new['pts'] - df_new['opp_pts']
df_new.head()
```

```python
# Create new column
df_new['selisih_poin'] = np.abs(df_new['pts'] - df_new['opp_pts'])  # selisihnya absolut
df_new.head()
```

```python
# rename column
df_new.rename(columns={'game_location' : 'loc',
                        'game_result': 'result'})   # perubahan tidak permanen (hanya overview)
```

```python
# rename column
df_new.rename(columns={'game_location' : 'loc',
                                'game_result': 'result'}, inplace=True)   # perubahan permanen
df_new
```

```python
# Drop column
df_new.drop('elo_i', axis=1)
```

```python
# Drop column
df_new.drop(['elo_i', 'elo_n'], axis=1, inplace=True) # permanen
```
